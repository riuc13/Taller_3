```python
import pandas as pd
```

```python
df = pd.read_excel("/content/gapminder.xlsx")
df=df[["lifeExp","pop","gdpPercap"]]
```

```python
import numpy as np

indices = np.random.permutation(len(df))
indices = indices[:170]

df["lifeExp"].loc[indices]=(pd.NA)
df["pop"].loc[indices]=(pd.NA)
df["gdpPercap"].loc[indices]=(pd.NA)

pd.isna(df["lifeExp"])
pd.isna(df["pop"])
pd.isna(df["gdpPercap"])
df
```

```python
#Importar el archivo gapminder en formato “xlsx”.
surveys = pd.read_excel("/content/gapminder.xlsx")
surveys
```

```python
#Graficar el diagrama de dispersión lifeEx vs pop.
import matplotlib.pyplot as plt
my_plot = surveys.plot("lifeExp", "pop", kind="scatter")
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
#Graficar el diagrama de dispersión gdpPercap vs pop. 

import matplotlib.pyplot as plt
my_plot = surveys.plot("gdpPercap", "pop", kind="scatter")
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
#Graficar los diagramas de cajas de la variable gdpPercap discrimina-
#dos por continentes desde 1990 a 2007.

surveys.filter(["year"])
surveys_filtrado = surveys[(surveys.year >= 1990)]
```

```python
import seaborn as sns
sns.set_style('whitegrid')
ax = sns.boxplot(x='continent',y='gdpPercap',data=surveys_filtrado)
ax = sns.stripplot(x="continent",y="gdpPercap",data=surveys_filtrado)
```
